## 绘制柱状图

```python
all_type = []
[all_type.extend(i.split()) for i in list(data['类型'])]
all_type
```

```python
from collections import Counter
```

```python
counter_type = Counter(all_type).items()  # 统计词频
counter_type = sorted(counter_type, key=lambda x: x[1], reverse=True)  # 按照出现次数进行从大到小的排序
counter_type =   counter_type[:10]  # 取出前十个类型绘制相应的图像
counter_type
```

```python
# 分离类型和次数
types = [i[0] for i in counter_type]
nums = [i[1] for i in counter_type]
```

```python
plt.figure(figsize=(6, 4), dpi=100)
# x：横坐标标签
# height：每根柱子的高度
# width：柱子的宽度，默认情况下等于0.8
# color：每根柱子的颜色
plt.bar(x=types, 
        height=nums, 
        width=0.5, 
        color=['#F7ADBC', '#DEAB8A', '#847E39', '#444793', '#B2D235', '#F05B72', '#2E3920', '#FAB37B', '#87481F'])
plt.show()
```

## 饼状图

```python
plt.figure(figsize=(6, 4), dpi=100)
# x：表示用来绘制饼状图的数据
# labels：表示每个数据对应的标签
# explode：表示每块扇形区域离中心点的距离
# autopct：表示显示每块扇形区域的格式
# colors：每块扇形区域的颜色
plt.pie(x=nums, 
        labels=types, 
        explode=[0.03]*10,
        autopct='%.1f%%',
        colors=['#F7ADBC', '#DEAB8A', '#847E39', '#444793', '#B2D235', '#F05B72', '#2E3920', '#FAB37B', '#87481F'])
plt.show()
```

### 小练习：
- 绘制comment_sina.xlsx中评论的词云图。注意：最好把表情符号想办法删除掉

```python
data4 = pd.read_excel('comment_sina.xlsx', index_col=0)
data4
```

```python
comment = data4['博文内容'].tolist()
```

```python
# 删除表情符号[？？？]
import re

print('替换前：',  data4['博文内容'][509])
print('替换后：', re.sub(r'\[.+?\]', '', data4['博文内容'][509]))
```

```python
comment = [re.sub(r'\[.+?\]', '', i) for i in comment]
```

```python
# 分词
cut_res = []
[cut_res.extend(jieba.lcut(i)) for i in comment if not pd.isna(i)]
cut_res
```

```python
# 去停用词
res = [i for i in cut_res if i not in stoplist]
```

```python
# 统计词频
count_word = Counter(res)
```

```python
# 绘制词云图
wc = WordCloud(font_path='simhei.ttf',  # 修改默认字体
              width=500,  # 设置词云图宽度
              height=500,  # 设置词云图高度
              background_color='#FFE793',  # 设置词云图背景色
              max_words=300 ,  # 设置最大显示的词语数量
              mask=bg)  # 背景模板图片

wc.fit_words(count_word)
```

```python
# 展示词云图
plt.figure(figsize=(3, 3), dpi=300)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
