### 翻页抓取

```python
url = 'https://weibo.com/ajax/statuses/buildComments?is_reload=1&id=4854095602783033&is_show_bulletin=2&is_mix=0&count=10&uid=2028810631'
allname = []
alltime = []
alltext = []
allgender = []
page = 1
n = 0

while True:
    res = requests.get(url)
    res_json = res.json()
    max_id = res_json['max_id']
    new = f'https://weibo.com/ajax/statuses/buildComments?is_reload=1&id=4854095602783033&is_show_bulletin=2&is_mix=0&max_id={max_id}&count=10&uid=2028810631'
    
    name = [i['user']['screen_name']for i in res_json['data']] 
    time=[i['created_at'] for i in res_json['data']]
    text=[i['text'] for i in res_json['data']]
    gender=[i['user']['gender'] for i in res_json['data']]
    
    if len(text) == 0:
        n+=1
    else:
        n = 0
        
    allname.extend(name)
    alltime.extend(time)
    alltext.extend(tmp_text)
    allgender.extend(gender)
    print(f'Page:{page},Total:{len(alltext)},Max_id:{max_id}')
    page+=1
    
    url = new
    if max_id == '0' or n>=10:
        print('Stop!')
        break
```

### 词云图（评论内容）

```python
pl = dq['评论内容'].to_list()
pl
```

```python
import jieba
cut_res = []
[cut_res.extend(jieba.lcut(i)) for i in pl if not pd.isna(i)]
cut_res
```

#### 去停用词

```python
# 读取停用词典
with open('stoplist.txt', encoding='UTF-8') as f:
    stoplist = f.readlines()
stoplist = [i.strip() for i in stoplist]
stoplist
```

```python
res = [i for i in cut_res if i not in stoplist]
```

```python
from collections import Counter
count_word = Counter(res)
len(count_word)
```

#### 绘制词云图

```python
from wordcloud import WordCloud
from PIL import Image
```

```python
wc = WordCloud(font_path='simhei.ttf',  # 修改默认字体
              width=500,  # 设置词云图宽度
              height=500,  # 设置词云图高度
              background_color='#FFE793',  # 设置词云图背景色
              max_words=300 ,  # 设置最大显示的词语数量
              )  # 背景模板图片
```

```python
wc.fit_words(count_word)
```
