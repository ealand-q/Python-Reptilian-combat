```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

# 折线图

```python
data=pd.read_excel('baidu_top30.xlsx',index_col=0)
```

```python
data
```

```python
plt.figure(figsize=(20,20),dpi=1090)
plt.plot(range(len(data['热度指数'])),data['热度指数'],"--.",color="red")
plt.xticks(range(len(data['热度指数'])),data['电影名称'],rotation=40)
plt.grid()
plt.show()
```
